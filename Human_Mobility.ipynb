```python
import pandas as pd
import numpy as np
import glob
from tqdm.contrib.concurrent import process_map
import os

# Paths
input_files = sorted(glob.glob('/n/netscratch/cga/Lab/rpalbuquerque/Mobility/Mobility Data (Splited)/mobility_part_*.csv.gz'))
output_path = '/n/netscratch/cga/Lab/rpalbuquerque/Mobility/Sisi/mobility_by_sector.csv.gz'

# Function to process a single file
def process_file(path):
    df = pd.read_csv(path, compression='gzip')
    grouped = df.groupby('store_id').agg(
        lambda x: np.sum(x) if np.issubdtype(x.dtype, np.number) else x.iloc[0]
    )
    return grouped

# Parallel processing with real-time progress bar
results = process_map(
    process_file,
    input_files,
    max_workers=64,
    chunksize=1,
    desc="üöÄ Processing mobility files"
)

# Concatenate all intermediate results
combined = pd.concat(results)

# Separate numeric and categorical columns
numeric_cols = combined.select_dtypes(include=np.number).columns
categorical_cols = [c for c in combined.columns if c not in numeric_cols]

# Final aggregation by store_id
mob_numeric = combined[numeric_cols].groupby(combined.index).sum()
mob_categ = combined[categorical_cols].groupby(combined.index).first()

# Merge numeric and categorical
mobility = pd.concat([mob_numeric, mob_categ], axis=1).reset_index().rename(columns={'index': 'store_id'})

# Save final result
mobility.to_csv(output_path, index=False, compression='gzip')
print(f"\n‚úÖ Final mobility file saved: {output_path} ‚Äî Total sectors: {mobility.shape[0]:,}")
```

```python
##üß± Block 2 ‚Äî Spatial Join (CD_SETOR + Geometry)
```

```python
import pandas as pd
import geopandas as gpd

# Paths
mobility_path = '/n/netscratch/cga/Lab/rpalbuquerque/Mobility/Sisi/mobility_by_sector.csv.gz'
shapefile_path = '/n/netscratch/cga/Lab/rpalbuquerque/e-WOM/BR_setores_CD2022.gpkg'
output_path = '/n/netscratch/cga/Lab/rpalbuquerque/Mobility/Sisi/mobility_by_sector_joined.gpkg'

# Load mobility data
mobility = pd.read_csv(mobility_path)
mobility['store_id'] = mobility['store_id'].astype(str)  # convert to string

# Load census tracts using pyogrio
gdf = gpd.read_file(shapefile_path, layer='BR_setores_CD2022', engine='pyogrio')
gdf['CD_SETOR'] = gdf['CD_SETOR'].astype(str)  # ensure it's string too

# Perform the merge
gdf_merged = gdf.merge(mobility, left_on='CD_SETOR', right_on='store_id', how='inner')

# Save as GeoPackage
gdf_merged.to_file(output_path, layer='mobility_by_sector', driver='GPKG')

print(f"‚úÖ Spatial join completed successfully: {gdf_merged.shape[0]:,} sectors matched and saved to:\n{output_path}")
```

```python
import geopandas as gpd
import pandas as pd
import numpy as np
from libpysal.weights import Queen
from esda.moran import Moran

# Load spatial data (force pyogrio)
gdf = gpd.read_file(
    '/n/netscratch/cga/Lab/rpalbuquerque/Mobility/Sisi/mobility_by_sector_joined.gpkg',
    layer='mobility_by_sector',
    engine='pyogrio'
)

# Fill NaNs (robustness)
gdf = gdf.fillna(0)

# Construct spatial weights matrix (Queen contiguity)
print("üì¶ Building spatial weights matrix...")
w = Queen.from_dataframe(gdf)
w.transform = 'r'  # row-standardized

# Define metrics
metrics = ['visits', 'unique', 'dwell_time_mins']

# Compute Moran's I for each metric
print("‚öôÔ∏è Computing Moran's I with 999 permutations...")
results = [
    {
        'Metric': var,
        "Moran's I": round((m := Moran(gdf[var].values, w, permutations=999)).I, 5),
        'p-value': round(m.p_sim, 5),
        'z-score': round(m.z_sim, 2)
    }
    for var in metrics
]

# Display result as table
results_df = pd.DataFrame(results)
display(results_df)
```

```python
##üß± Block 5 ‚Äì LISA Maps (with Pyogrio, Dark Style, PNG Output)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from libpysal.weights import Queen
from esda.moran import Moran_Local
import numpy as np
import os

# Load GeoDataFrame (safe loading)
gdf = gpd.read_file(
    '/n/netscratch/cga/Lab/rpalbuquerque/Mobility/Sisi/mobility_by_sector_joined.gpkg',
    layer='mobility_by_sector',
    engine='pyogrio'
)
gdf = gdf.fillna(0)

# Spatial weights
w = Queen.from_dataframe(gdf)
w.transform = 'r'

# Output path
output_path = '/n/netscratch/cga/Lab/rpalbuquerque/Mobility/Sisi/maps/lisa/'
os.makedirs(output_path, exist_ok=True)

# Style: dark mode
plt.style.use('dark_background')

# Metrics to analyze
metrics = ['visits', 'unique', 'dwell_time_mins']

# Cluster labels
cluster_labels = {
    1: "High-High",
    2: "Low-Low",
    3: "Low-High",
    4: "High-Low"
}

for col in metrics:
    x = gdf[col].values
    lisa = Moran_Local(x, w, permutations=999)

    # Create LISA cluster column
    gdf['lisa_cluster'] = lisa.q
    gdf['lisa_sig'] = lisa.p_sim < 0.05

    # Apply labels only to significant results
    gdf['cluster_type'] = np.where(
        gdf['lisa_sig'],
        gdf['lisa_cluster'].map(cluster_labels),
        'Not Significant'
    )

    # Plot
    fig, ax = plt.subplots(figsize=(12, 10))
    gdf.plot(column='cluster_type',
             categorical=True,
             legend=True,
             cmap='Set1',
             linewidth=0.05,
             edgecolor='black',
             ax=ax)

    ax.set_title(f"LISA Cluster Map: {col.replace('_', ' ').title()}", fontsize=16, pad=20)
    ax.axis('off')
    plt.tight_layout()

    # Save map
    map_path = os.path.join(output_path, f'lisa_{col}.png')
    plt.savefig(map_path, dpi=300, bbox_inches='tight')
    plt.close()

    print(f"‚úÖ LISA map saved for {col}: {map_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
from IPython.display import display

# Load data
mobility_path = '/n/netscratch/cga/Lab/rpalbuquerque/Mobility/Sisi/mobility_by_sector.csv.gz'
df = pd.read_csv(mobility_path)

# Set aesthetic style
sns.set(style="whitegrid")
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

# Plot helper
def plot_bar(data, column, title):
    order = data[column].value_counts().index
    plt.figure(figsize=(8,4))
    ax = sns.countplot(data=data, x=column, order=order, palette="viridis")
    plt.title(title)
    plt.ylabel("Number of Sectors")
    plt.xlabel("")
    plt.xticks(rotation=45)
    for p in ax.patches:
        ax.annotate(f'{p.get_height():,}', (p.get_x()+0.15, p.get_height()+1000), fontsize=9)
    plt.tight_layout()
    plt.show()

# Gender distribution
plot_bar(df, 'demographics_gender', 'Distribution by Gender')

# Age range distribution
plot_bar(df, 'demographics_age_range', 'Distribution by Age Range')

# Social class distribution
plot_bar(df, 'demographics_class', 'Distribution by Social Class')

# State coverage
plot_bar(df, 'state', 'Number of Census Tracts per State')

# Summary table
summary = df[['demographics_gender', 'demographics_age_range', 'demographics_class', 'state']].describe(include='all')
display(summary)
```
